```python
import pandas as pd
```

Downloading and merging EXIST test and training data set

```python


# Load the training and test files
train_path = "/content/EXIST2021_training.tsv (1).txt"
test_path = "/content/EXIST2021_test.tsv (1).txt"

# Read the TSV files
df_train = pd.read_csv(train_path, sep='\t')
df_test = pd.read_csv(test_path, sep='\t')

# Add missing task columns to the test set
df_test['task1'] = pd.NA
df_test['task2'] = pd.NA

# Add a column to indicate dataset source
df_train['split'] = 'train'
df_test['split'] = 'test'

# Combine datasets
df_combined = pd.concat([df_train, df_test], ignore_index=True)

# Save to CSV or display a preview
df_combined.to_csv("EXIST2021_combined.csv", index=False)

# Optional: show first few rows
print(df_combined.head())
```
