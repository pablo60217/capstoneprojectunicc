```python
import requests
import pandas as pd
import time
import re

def run_ollama(prompt, model="Mistral"):
    response = requests.post(
        "http://localhost:11434/api/generate",
        json={"model": model, "prompt": prompt, "stream": False}
    )
    if response.ok:
        return response.json()["response"]
    return "no response"




for index, row in df.iterrows():
    tweet = row["text"]
    

    prompt = f"""
   You are a classifier that labels tweets as either **sexist** or **non-sexist**.

    **Definition:**
    - A tweet is **sexist** if it includes:
    - Discrimination, stereotyping, or exclusion based on gender or sex.
    - Harmful generalizations (e.g. "women are bad drivers", "men can't be emotional").
    - Objectification or sexualization based on gender.
    - Dismissive or condescending comments targeting a gender.
    - Blaming or justifying violence or inequality toward a gender.

    - A tweet is **non-sexist** if it:
    - Expresses a neutral or respectful opinion.
    - Does not refer to gender at all.
    - Discusses gender issues in a respectful and factual way.
    - Contains jokes, critiques, or opinions that are not derogatory toward a gender.

    **Examples of sexist tweets:**
    - "Women should stay in the kitchen."
    - "Men can't raise kids properly."
    - "She was asking for it by dressing like that."

    **Examples of non-sexist tweets:**
    - "I love working with my female colleagues."
    - "Gender equality is important in every society."
    - "He‚Äôs a great dad who raised his kids alone."

    Now classify the following tweet:

    "{tweet}"

    Respond with only one word: sexist or non-sexist.
    """.strip()

    print(f"‚Üí Row {index}: sending to model...")
    answer = run_ollama(prompt)
    print(answer)

    # Extract 'sexist' or 'non-sexist' from the response
    
    Type = "unknown"
    match = re.search(r"\b(sexist|non-sexist)\b", answer.lower())
    if match:
        Type = match.group(1)

    df.at[index, "Type"] = Type
    time.sleep(3)

df
```

```python
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, confusion_matrix, classification_report
)

y_true = df["real_label"]
y_pred = df["Type"]

# Accuracy general
accuracy = accuracy_score(y_true, y_pred)

# Precision, Recall, F1 por clase (promedio macro)
precision = precision_score(y_true, y_pred, pos_label="sexist", average='binary')
recall = recall_score(y_true, y_pred, pos_label="sexist", average='binary')
f1 = f1_score(y_true, y_pred, pos_label="sexist", average='binary')

# Matriz de confusi√≥n
conf_matrix = confusion_matrix(y_true, y_pred, labels=["sexist", "non-sexist"])

# Reporte completo
report = classification_report(y_true, y_pred, labels=["sexist", "non-sexist"])

# Imprimir resultados
print(f"üîç Accuracy:  {accuracy:.2%}")
print(f"üéØ Precision: {precision:.2%}")
print(f"üìà Recall:    {recall:.2%}")
print(f"üìä F1 Score:  {f1:.2%}")
print("\nüßÆ Confusion Matrix:")
print(conf_matrix)
print("\nüìã Classification Report:")
print(report)

```
